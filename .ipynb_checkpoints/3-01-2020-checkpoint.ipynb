# how to group by method in dataframe

```python
import pandas as pd
import numpy as np
```

```python
f=pd.DataFrame({'A':['foo','bar','foo','bar',
                    'foo','bar','foo','foo'],
            'B':['one','one','two','three',
                'two','two','one','three'],
               'C':np.random.randint(8),'D':np.random.randint(8)})
f
```

```python
f.groupby('A').sum()
```

```python
f.groupby(['A','B']).sum()
```

#### RANK METHOD IN DATAFRAME

```python
df=pd.DataFrame(data={'names':['gopi','sana','asim','dhathu','hari'],
                     'marks':[99,100,55,55,np.nan]},dtype=int)
df
```

### default rank
### max_rank
###  NA_bottom
### NA_top

```python
df['defaults_rank']=df["marks"].rank()
df      #Defaults rank in selected in botttom 2+1=3......3/2=1.5 is value is given bottom
        
```

```python
    df['max_rank']=df['marks'].rank(method='max')
    df     # its take 2 because method called maximum values
```

```python
df['min_rank']=df['marks'].rank(method='min')
df     # its take 1 because method called minimum values
```

```python
df['NA_bottom']=df['marks'].rank(na_option='bottom')
df
```

```python
df['NA_top']=df['marks'].rank(na_option='top')
df.index=[1,2,3,4,5]
df
```

```python
df['pct_rank']=df['marks'].rank(pct=True)
df
```
