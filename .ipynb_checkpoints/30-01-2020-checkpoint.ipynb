```python
print('B/n',np.linspace(2.0,3.0,num=5),'\n')
x=np.linspace(0,2,10)
print('A/n',np.sin(x))
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range (1,5):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sinplot()
plt.show()
```

```python
print('B/n',np.linspace(2.0,3.0,num=5),'\n')
x=np.linspace(0,2,10)
print('A/n',np.sin(x))
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range (1,5):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sns.set() # default style
sinplot()
plt.show()
```

```python
# styles of seaborn:-
# whitegrid
# darkgrid
# dark
#white
#ticks
print('B/n',np.linspace(2.0,3.0,num=5),'\n')
x=np.linspace(0,2,10)
print('A/n',np.sin(x))
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range (1,50): # you increase the range of plot you feel very messy
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sns.set_style("whitegrid") # you can changes style
sinplot()
plt.show()
```

```python
#plot a basic univariate density(to call x and y)
np.random.seed(10) # seed funtionlity to help the random value is fix but,
                   #randn changed values every time to run the process
sns.set(color_codes=True)
a,c =[0,2],[(1,0.5),(0.5,1)]
x,y=np.random.multivariate_normal(a,c,size=50).T
ax=sns.kdeplot(x)
```

```python
np.random.seed(10) 
sns.set(color_codes=True)
a,c =[0,2],[(1,0.5),(0.5,1)]
x,y=np.random.multivariate_normal(a,c,size=50).T
# ax=sns.kdeplot(x)
ay=sns.kdeplot(x)
```

```python
np.random.seed(10) 
sns.set(color_codes=True)
a,c =[0,2],[(1,0.5),(0.5,1)]
x,y=np.random.multivariate_normal(a,c,size=50).T

ay=sns.kdeplot(x,shade=True,color='b')   # SHADE IS USED TO TO FILL THE SHADE IN PLOT
ax=sns.kdeplot(y)
```

```python
np.random.seed(10) 
sns.set(color_codes=True)
a,c =[0,2],[(1,0.5),(0.5,1)]
x,y=np.random.multivariate_normal(a,c,size=50).T

ay=sns.kdeplot(x,y)
```

```python
iris=sns.load_dataset('iris')
print(iris)
ax=iris.loc[iris.species=='setosa']
ay=iris.loc[iris.species=='virginica']
a=sns.kdeplot(ax.sepal_width,ax.sepal_length,shade=True,c='r',shade_lowest=False)
y=sns.kdeplot(ay.sepal_width,ay.sepal_length,shade=True,c='k',shade_lowest=False)
```
