```python
x=[3,2,1,5,6]
y=[4,5,6,7,8]
x1=[5,4,3,2,1]
y1=[9,8,7,6,5]
plt.plot(x,y,'r',ls='dashed',linewidth=8,label='first element')
plt.plot(x1,y1,'b',linewidth=2,label='second element')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```
