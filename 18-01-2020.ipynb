# scatter plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
```

```python
l1=[1,2,3,3,4,44,4]
l2=[3,5,6,6,3,3,33]
plt.scatter(l1,l2,color='r')
plt.show()
```

```python
days=[1,2,3,4,5,6,7]
sleep=[5,6,6,7,8,7,9]
eating=[1,1,3,2,1,4,3]
working=[9,8,8,7,6,12,2]
playing=[1,3,4,5,4,5,5]
co=['y','g','b','k']
plt.stackplot(days,sleep,eating,working,playing,colors=co)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Day activities")
plt.show()
```

```python
plt.pie(sleep,labels=['mon','tue','wed','thu','fri','sat','sun'],autopct="%.3f")
plt.title("Daily sleeping")
plt.show()
```

```python
def f(t):
    return np.exp(-t)*np.sin(2*np.pi*t)
d1=np.arange(0.0,0.5,0.01)
d2=np.arange(0.0,5.0,0.01)
plt.subplot(221)  # middle val 0.5=alignment,vertical,horizontal
plt.plot(d1,f(d1),"go",d2,f(d2))
#---------------------------------------
plt.subplot(222)   #  middle val 0.5=aligment,vertical,horizontal 2
plt.plot(d2,np.sin(2*np.pi*d2),'yo')
```
