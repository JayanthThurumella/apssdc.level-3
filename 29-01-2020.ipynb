```python
from matplotlib import style
style.use('seaborn-notebook')
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
x=[1,2,3,4,5,6,7,8,9]
y=[9,5,3,8,3,1,0,6,7]
z=[7,4,5,7,2,5,7,1,8]
ax1.plot(x,y,z,color='orange')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')
plt.show()
```

```python
import numpy as np
style.use('grayscale')
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
a=[9,4,5,6,8,8,9,5,7,8]
b=[1,3,4,5,7,8,8,5,1,1]
da=np.ones(10)
db=np.ones(10)
dc=[2,3,4,5,6,7,8,8,5,3]
c=np.zeros(10)

dc=[1,2,3,4,5,10,6,5,9,1]
ax1.bar3d(a,b,c,da,db,dc,color='orange')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')
plt.show()
```

```python
style.use('classic')
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
x=[8,3,3,3,2,2,4]
y=[1,5,5,9,6,4,6]
z=[1,3,6,8,7,7,7]

ax=[-1,-2,-3,-4,-5,-6,-7]
ay=[-5,-8,-3,-9,-1,-6,-4]
az=[5,6,7,8,9,3,6]
ax1.scatter(x,y,z,c='r',marker='o')
ax1.scatter(x,y,z,c='k',marker='o')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')
plt.show()

            

```

```python
style.use('classic')
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
x=[8,3,3,3,2,2,4]
y=[1,5,5,9,6,4,6]
z=[1,3,6,8,7,7,7]

ax=[-1,-2,-3,-4,-5,-6,-7]
ay=[-5,-8,-3,-9,-1,-6,-4]
az=[5,6,7,8,9,3,6]
ax1.scatter(x,y,z,c='r',marker='o')
ax1.scatter(ax,ay,az,c='k',marker='o')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')
plt.show()
```
