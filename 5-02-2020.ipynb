```python
import pandas as pd
import numpy as np
```

```python
pd.date_range('2019-01-01','2020-01-01',freq='D')  # DAY
```

```python
pd.date_range('2019-01-01','2020-01-01',freq='2D') # 2DAYS VARIATION
```

```python
pd.date_range('2019-01-01','2020-01-01',freq='3W') #3 WEEK WARIATION
```

```python
pd.date_range('2019-01-01','2022-01-01',freq='Y') #Y MEAN YEAR
```

```python
pd.date_range('2019-01-01','2020-01-01',freq='M') # M MEANS MONTH
```

```python
pd.date_range('2019-01-01','2020-01-01',freq='T') # T MEANS MINs
```

```python
pd.date_range('2019-01-01','2020-01-01',freq='S')   # SECONDS
```

```python
dates=pd.date_range('1/1/2019',periods=100,freq='W-SAT') # SAT MEANS SATURDAY
dates                                                    # 100 PARTS DIVDED
```

```python
a=pd.DataFrame(np.random.randn(100,4 ),index=dates,columns=['colorado','Texas','newyork','ohio'])
a                             #100 rows and 4 columns
```

```python
a.loc["2020-11-28"] #  or 2020/11/28 are used
```
