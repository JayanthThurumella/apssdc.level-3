```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.pyplot import style
```

```python
a1=plt.subplot(2,2,1)
a2=plt.subplot(2,2,2)
a3=plt.subplot(2,2,3)
a4=plt.subplot(2,2,4)
```

#  ------------------------     OR      ------------------------------------

```python
a=plt.figure()
a1=a.add_subplot(2,2,1)
a2=a.add_subplot(2,2,2)
a3=a.add_subplot(2,2,3)
a4=a.add_subplot(2,2,4)
```

```python
a1=plt.subplot(2,2,1)
a2=plt.subplot(2,2,2)
a3=plt.subplot(2,2,3)
a4=plt.subplot(2,2,4)
days=['mon','tue','wed','thurs','fri','sat','sun']
sleep=[7,7,7,7,6,6,10]
a3.plot(np.random.randn(50).cumsum(),'b--')
a1.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
a2.hist(np.random.randn(100),color='k',orientation='horizontal')
a4.pie(sleep,labels=days,autopct='%0.2f')
plt.show()
```

```python
x=[5,10,15,20,25,30,35,40,45,50]
y=[1,4,3,2,7,6,9,8,10,5]
plt.plot(x,y,'.')
plt.xlabel("x list values")
plt.ylabel("y list values")
plt.xticks(np.arange(0,51,5))
plt.yticks(np.arange(0,11,1))
plt.show()
```
